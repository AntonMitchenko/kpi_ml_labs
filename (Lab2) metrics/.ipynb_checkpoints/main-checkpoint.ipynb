## 3. Для зчитаного набору даних виконати наступні дії:

a. Обчислити всі метрики (Accuracy, Precision, Recall, F-Scores,
Matthews Correlation Coefficient, Balanced Accuracy, Youden’s J
statistics, Area Under Curve for Precision-Recall Curve, Area
Under Curve for Receiver Operation Curve) для кожної моделі при
різних значеннях порогу класифікатора (крок зміни порогу
обрати самостійно).

```python
# Спочатку зробимо трешхолди та функцію яка буде розбивати на перший або другий клас.
classificators = np.linspace(0.01, 0.99, 9)

def one_or_zero(l,clasificator):
    tmp_l = []
    clasificator_score1 = 0
    clasificator_score0 = 0
    for i in l:
        if i > clasificator:
            clasificator_score1 += 1
            tmp_l.append(1)
        else:
            tmp_l.append(0)
            clasificator_score0 += 1
    one_score.append(clasificator_score1)
    zero_score.append(clasificator_score0)
    return tmp_l
```

```python
# Потім функцію яка буде вираховувати усі потрібні метрики.
def metrics_func(tr_data, mass):        
    Accuracy = accuracy_score(tr_data, mass)
    Precision = precision_score(tr_data, mass, average='binary')
    Recall = recall_score(tr_data, mass, average='binary')
    F_Scores = f1_score(tr_data, mass, average='binary')
    MCC = matthews_corrcoef(tr_data, mass)
    BalancedAcc = balanced_accuracy_score(tr_data, mass)
    YJstat = roc_auc_score(tr_data, mass)
    precision, recall, thresholds = precision_recall_curve(tr_data, mass)
    AUC_PR = auc(precision, recall)
    fpr, tpr, thresholds = roc_curve(tr_data, mass, pos_label=1)
    AUC_ROC = auc(fpr, tpr)
    
    return Accuracy,Precision,Recall,F_Scores,MCC,BalancedAcc,YJstat,AUC_PR,AUC_ROC,fpr,tpr
```

```python
model1_metrics = []
one_score = []
zero_score = []
for i in classificators:
    predict = one_or_zero(df["Model_1"],i)
    metrics = metrics_func(df["GT"],predict)
    model1_metrics.append(metrics)
```

b. Збудувати на одному графіку в одній координатній системі
(величина порогу; значення метрики) графіки усіх обчислених
метрик, відмітивши певним чином максимальне значення кожної
з них.

```python
accuracy = [model1_metrics[i][0] for i in range(9)]
precision = [model1_metrics[i][1] for i in range(9)]
recall = [model1_metrics[i][2] for i in range(9)]
f_scores = [model1_metrics[i][3] for i in range(9)]
mcc = [model1_metrics[i][4] for i in range(9)]
balancedacc = [model1_metrics[i][5] for i in range(9)]
yjstat = [model1_metrics[i][6] for i in range(9)]
auc_pr = [model1_metrics[i][7] for i in range(9)]
auc_roc = [model1_metrics[i][8] for i in range(9)]
```

```python
# Потім функцію яка буде вираховувати усі потрібні метрики.
def plot_metrics(text, accuracy, precision, recall, f_scores, mcc, balancedacc, yjstat, auc_pr, auc_roc):
    plt.plot(classificators, accuracy, label ='accuracy') 
    plt.plot(classificators, precision, label ='precision')
    plt.plot(classificators, recall, label ='recall')
    plt.plot(classificators, f_scores, label ='f_scores')
    plt.plot(classificators, mcc, label ='mcc')
    plt.plot(classificators, balancedacc, label ='balancedacc')
    plt.plot(classificators, yjstat, label ='yjstat')
    plt.plot(classificators, auc_pr, label ='auc_pr')
    plt.plot(classificators, auc_roc, label ='auc_roc')

    plt.title(f" {text} metrics")
    plt.xlabel("classificator")
    plt.ylabel("Score")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_metrics("Model 1", accuracy, precision, recall, f_scores, mcc, balancedacc, yjstat, auc_pr, auc_roc)
```

c. Збудувати в координатах (значення оцінки класифікаторів;
кількість об’єктів кожного класу) окремі для кожного класу
графіки кількості об’єктів та відмітити вертикальними лініями
оптимальні пороги відсічення для кожної метрики.

```python
def c_function(one_score, zero_score, accuracy, precision, recall, f_scores, auc_roc):    
    plt.plot(classificators, one_score, label ='1') 
    plt.plot(classificators, zero_score, label ='0')
    plt.axvline(classificators[np.argmax(accuracy)], color='red', linestyle='--')
    plt.text(classificators[np.argmax(accuracy)], 0.9, f'accuracy: {classificators[np.argmax(accuracy)]}', color='red', rotation=90, verticalalignment='bottom')

    plt.axvline(classificators[np.argmax(precision)], color='blue', linestyle='--')
    plt.text(classificators[np.argmax(precision)], 0.9, f'precision: {classificators[np.argmax(precision)]}', color='blue', rotation=90, verticalalignment='bottom')

    plt.axvline(classificators[np.argmax(recall)], color='green', linestyle='--')
    plt.text(classificators[np.argmax(recall)], 0.9, f'recall: {classificators[np.argmax(recall)]}', color='green', rotation=90, verticalalignment='bottom')

    plt.axvline(classificators[np.argmax(f_scores)], color='grey', linestyle='--')
    plt.text(classificators[np.argmax(f_scores)], 0.9, f'accuracy: {classificators[np.argmax(f_scores)]}', color='grey', rotation=90, verticalalignment='bottom')

    plt.axvline(classificators[np.argmax(auc_roc)], color='orange', linestyle='--')
    plt.text(classificators[np.argmax(auc_roc)], 0.9, f'auc_roc: {classificators[np.argmax(auc_roc)]}', color='orange', rotation=90, verticalalignment='bottom')

    plt.title(f" Zero and one score")
    plt.xlabel("classificator")
    plt.ylabel("Score")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
c_function(one_score, zero_score, accuracy, precision, recall, f_scores, auc_roc)
```

## 5. Створити новий набір даних, прибравши з початкового набору
(50 + 10К)% об’єктів класу 1, вибраних випадковим чином. Параметр К
представляє собою залишок від ділення місяця народження студента на
чотири та має визначатися в програмі на основі дати народження
студента, яка задана в програмі у вигляді текстової змінної формату
‘DD-MM’.

```python
# Mitchenko Anton var 11  date birthday  = 30 - 06
k = 6 % 4

count_of_1 = df.query("GT == 1")['GT'].count()
```

## 6. Вивести відсоток видалених об’єктів класу 1 та кількість елементів кожного класу після видалення.

```python
delet_elem_clas1 = int(count_of_1 - count_of_1 * (50 + 10 * k) / 100)
new_df = pd.concat([df.query("GT == 1").sample(n=delet_elem_clas1,random_state=1, ignore_index=True), df.query("GT == 0")], ignore_index=True)
print((delet_elem_clas1*100)/5000, f'% - Відсоток віддалених одиниць від загальної кількості    {50 + 10 * k}% - Відсоток віддалених одиниць від одиниць.')
print(new_df["GT"].value_counts())  # not balansed
```
