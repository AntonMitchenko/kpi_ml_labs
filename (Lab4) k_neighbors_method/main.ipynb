# 4. Отримати десять варіантів перемішування набору даних
та розділення його на навчальну (тренувальну) та тестову вибірки, використовуючи функцію ShuffleSplit. Сформувати начальну та тестові вибірки наоснові восьмого варіанту. З’ясувати збалансованість набору даних.

```python
# Split data
y = df["quality"]
X = df.drop(["quality"], axis = 1)
shuffle_split = ShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
```

```python
y.value_counts(dropna=False).sort_index()/len(df)
```

```python
my_variant = 8
```

```python
for i, (train_index, test_index) in enumerate(shuffle_split.split(X)):
    
    if i==my_variant-1:
        print(f"Fold {i}:")
        print(f"  Train: index={train_index[0]}:{train_index[-1]}")
        print(f"  Test:  index={test_index[0]}:{test_index[-1]}")

        X_train, X_test = X.loc[train_index], X.loc[test_index]
        y_train, y_test = y.loc[train_index], y.loc[test_index] 
```

# 5. Використовуючи функцію KNeighborsClassifier
бібліотеки scikit-learn, збудувати класифікаційну модель на основі методу k найближчих сусідів (значення всіх параметрів залишити за замовчуванням) та навчити її на тренувальній вибірці, вважаючи, що цільова характеристика визначається стовпчиком quality, а всі інші виступають в ролі вихідних аргументів.

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
accuracy = knn.score(X_test, y_test)
print("Accuracy:", accuracy)
```

# 6. Обчислити класифікаційні метрики збудованої моделі
для тренувальної та тестової вибірки. Представити результати роботи моделі на тестовій вибірці графічно.

```python
def print_metrics(y_test, y_pred, y_proba=None):
    Accuracy = accuracy_score(y_test, y_pred).round(3)
    Precision = precision_score(y_test, y_pred, average='macro').round(3)
    Recall = recall_score(y_test, y_pred, average='macro').round(3)
    F_Scores = f1_score(y_test, y_pred, average='macro').round(3)
    MCC = matthews_corrcoef(y_test, y_pred).round(3)
    BalancedAcc = balanced_accuracy_score(y_test, y_pred).round(3)
    print("Accuracy = ", Accuracy)
    print("Precision =", Precision)
    print("Recall = ", recall_score(y_test, y_pred, average='macro').round(3))
    print("F_Scores = ", f1_score(y_test, y_pred, average='macro').round(3))
    print("MCC = ", matthews_corrcoef(y_test, y_pred).round(3))
    print("BalancedAcc = ", balanced_accuracy_score(y_test, y_pred).round(3))

    # Help here
    if y_proba is not None:
        roc_auc_stat = roc_auc_score(y_test, y_proba, multi_class='ovr')
        print('ROC-AUC Stat:', roc_auc_stat.round(3))

    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred).round(3))
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred).round(3))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)).round(3))
    return Accuracy, Precision, Recall, F_Scores, MCC, BalancedAcc, roc_auc_stat
```

```python
Accuracy_list = []
Accuracy_test_list = []
Precision_list = [] 
Precision_test_list = []
Recall_list = []
Recall_test_list = []
F_Scores_list = [] 
F_Scores_test_list = []
MCC_list = []
MCC_test_list = []
BalancedAcc_list = []
BalancedAcc_test_list = []
RoC_AUC = []
RoC_AUC_test = []


for i in range(20):
    print(f"\n Number of neighbors = {i+1}")
    
    knn = KNeighborsClassifier(n_neighbors=i+1)
    knn.fit(X_train, y_train)
    y_proba = knn.predict_proba(X_test)
    y_pred = knn.predict(X_test)
    y_proba_train = knn.predict_proba(X_train)
    y_pred_train = knn.predict(X_train)
    
    print(f"\n Train")
    Accuracy_test, Precision_test, Recall_test, F_Scores_test, MCC_test, BalancedAcc_test, roc_auc_stat_test = print_metrics(y_train, y_pred_train, y_proba_train)
    print(f"\n Test")
    Accuracy, Precision, Recall, F_Scores, MCC, BalancedAcc, roc_auc_stat = print_metrics(y_test, y_pred, y_proba)
    
#     Train
    Accuracy_test_list.append(Accuracy_test)
    Precision_test_list.append(Precision_test)
    Recall_test_list.append(Recall_test)
    F_Scores_test_list.append(F_Scores_test)
    MCC_test_list.append(MCC_test)
    BalancedAcc_test_list.append(BalancedAcc_test)
    RoC_AUC_test.append(roc_auc_stat_test)
    
    
#     Test
    Accuracy_list.append(Accuracy)
    Precision_list.append(Precision)
    Recall_list.append(Recall)
    F_Scores_list.append(F_Scores)
    MCC_list.append(MCC)
    BalancedAcc_list.append(BalancedAcc)
    RoC_AUC.append(roc_auc_stat)

    print("\n ---------------------------------")
```
